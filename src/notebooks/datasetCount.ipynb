```python
from pathlib import Path
import pandas as pd

root = Path(r"C:\Users\tim12\Desktop\deepfakeAI_project\data\raw") #데이터셋 폴더 경로

# 이미지, 영상 확장자 정의
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}
VID_EXTS = {".mp4", ".avi", ".mov", "mkv", ".wmv", ".flv"}

rows = [] #결과 담는 리스트

for dataset_dir in sorted([d for d in root.iterdir() if d.is_dir()]): #데이터셋 폴더
    dataset_name = dataset_dir.name
    row = {"dataset": dataset_name}

    # real, fake 클래스별 카운트
    for cls in ("real", "fake"):
        img_n = sum(
            1 for p in dataset_dir.rglob("*")
            if p.is_file()
            and p.suffix.lower() in IMG_EXTS
            and any(part.lower() == cls for part in p.parts)
        )

        vid_n = sum(
            1 for p in dataset_dir.rglob("*")
            if p.is_file()
            and p.suffix.lower() in VID_EXTS
            and any(part.lower() == cls for part in p.parts)
        )
        row[f"{cls}_images"] = img_n
        row[f"{cls}_videos"] = vid_n
        row[f"{cls}_total"] = img_n + vid_n

    row["dataset_total"] = row.get("real_total", 0) + row.get("fake_total", 0)
    rows.append(row)

#표 만들기
df = pd.DataFrame(rows).set_index("dataset").fillna(0).astype(int)
df
```
